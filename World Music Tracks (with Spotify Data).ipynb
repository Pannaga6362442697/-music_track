```python
#Popularity over the time
sns.relplot(x="Year",y="Popularity",kind='line',data=data)
```

```python
#Danceability Count
plt.figure(figsize=(10,6))
sns.lineplot(x='Year',y='Danceability',data=data,color='red')
plt.title('Trends in Danceability over time')
plt.show()
```

```python
# Sorting the data by Popularity for the top 10 tracks
top_10_data = data.nlargest(10, 'Popularity')

sns.barplot(x='Popularity', y='Track', data=top_10_data)
```

```python
#Number of listeners
plt.figure(figsize=(10,6))
plt.hist([data['Tempo'],data['Popularity']])
plt.xlabel('Count')
plt.ylabel('No.of listeners')
plt.legend(['Tempo'])
plt.show()
```

```python
#Maximum number in each key
y=['Popularity','Tempo']
width=data['Popularity'].max(),data['Tempo'].max()
plt.figure(figsize=(10,3))
plt.barh(y=y,width=width)
plt.title('Maximum number in each key')
for i,v in enumerate(width):
    plt.text(v,i," "+str(round(v)),color='green',va='center',fontweight='bold')
plt.show()
```

```python
#Number of songs released per year
top = data['Year'].value_counts().nlargest(10).index
top_data = data[data['Year'].isin(top)]

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Year', data=top_data)
plt.title("Number of songs released per year")

for i,v in enumerate(ax.patches):
    height = v.get_height()
    ax.text(v.get_x() + v.get_width() / 2., height + 0.5, 
            f'{int(height)}', ha="center", va='center', 
            fontweight='bold', color='green')

plt.show()
```
